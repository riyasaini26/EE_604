```python
imge = cv2.imread('jigsaw.jpg')
```

```python
#to divide the image in five parts and then divide it accordingly with given jigsaw pieces and arranging it and clubing it
pa1 = imge[0:200,0:190]
pa2 = imge[200:410,0:190]
pa3 = imge[410:421,0:190]
pa2 = cv2.flip(pa2,0)
piece1 = np.vstack((pa2,pa1,pa3))
```

```python
piece2 = imge[0:421,190:373]
```

```python
pa4 = imge[369:421,373:516]
pa5 = imge[0:369,373:516]
pa4 = cv2.flip(pa4,0)
piece3 = np.vstack((pa5,pa4))
```

```python
pa6 = imge[0:150,516:700]
pa7 = imge[150:348,516:700]
pa7 = cv2.flip(pa7,1)
pa8 = imge[348:370,516:700]
pa9 = imge[370:421,516:700]
pa9 = cv2.flip(pa9,0)
piece4 = np.vstack((pa6,pa7,pa8,pa9))
```

```python
piece5 = imge[0:421,700:797]
```

```python
solve = np.hstack((piece1,piece2,piece3,piece4,piece5))
```

```python
cv2.imshow('solved_puzzle',solve)
cv2.waitKey(0)
```
